```python
import json
import pandas as pd
import numpy as np
from datetime import datetime , timedelta

import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.dates import MO, TU, WE, TH, FR, SA, SU, DAILY, HOURLY, RRuleLocator,DateFormatter
import matplotlib.dates as mdates

sns.set(palette = 'muted', style = 'ticks')
```

```python
with open('logins.json') as outfile:
    login_data = json.load(outfile)
```

```python
times = login_data['login_time']
logins = pd.DataFrame(np.ones((len(times))) , index = pd.to_datetime(times), columns = ['freq'])
```

```python
logins_by15 = logins.resample(timedelta(minutes = 15)).sum()
```

```python
def plot_login_oneweek(df, ax, title, date_range = (), wseq = (MO, TU, WE, TH, FR, SA, SU), column = 'freq'):
    
    week = df[date_range[0]:date_range[1]]
    
    #Formats and locators used for major and minor ticks
    weekdays = mdates.WeekdayLocator(wseq)
    hours = mdates.HourLocator(byhour = [3,12,21])
    day_fmt = DateFormatter("%a")
    hour_fmt = DateFormatter("%-I:%M")
    
    #Set tick labels and prameters
    ax.xaxis.set_major_locator(weekdays)
    ax.xaxis.set_major_formatter(day_fmt)
    
    ax.xaxis.set_minor_locator(hours)
    ax.xaxis.set_minor_formatter(hour_fmt)
    
    ax.xaxis.set_tick_params(length=25)
    
    #Set title and labels
    ax.set_title(title, fontsize  = 18)
    ax.set_xlabel("Date" , fontsize  = 14)
    ax.set_ylabel("Login Count", fontsize  = 14)
    
    #Plot login counts
    x = week.index.to_pydatetime()
    y = week[column]
    ax.set_xlim(left = x.min(), right = x.max())
    ax.plot(x, y)
    
    #Get postions of dats and plot vertical lines for daily intevals
    day_positions = pd.to_datetime(pd.date_range(date_range[0],date_range[1]))
    for xc in day_positions:
        ax.axvline(x=xc, color='k', linestyle='dashed')
        
    #daily_means = logins_by15[date_range[0]:date_range[1]].resample(timedelta(days = 1)).mean()
    
    #for m in daily_means.itertuples(name=None):
        #left = pd.to_datetime(m[0])
        #right = pd.to_datetime(m[0] + pd.to_timedelta(1, unit = 'D'))
        #ax.axhline(y = m[1], xmin = left, xmax = right , color='r', linestyle='dashed')
```

```python
_ = f1, ax1 = plt.subplots(figsize=(15, 5))
plot_login_oneweek(logins_by15, ax1, "Login Frequency January 12th - 19th 1970" , 
                   date_range = ('1970-01-12','1970-01-19'))
```

```python
_ = f2, ax2 = plt.subplots(figsize=(15, 5))
plot_login_oneweek(logins_by15, ax2, "Login Frequency April 12th - 19th 1970" , 
                   date_range = ('1970-04-06','1970-04-13'))
```

```python
logins_by60 = logins_by15.resample(timedelta(hours = 1)).sum()
logins_daily_mean = logins_by60.resample(timedelta(days = 1)).mean()
logins_daily_mean.columns = ['hourly_average']
```

```python
avg_dailymean_bymonth = logins_daily_mean.resample('M').mean()
print(avg_dailymean_bymonth)
```
